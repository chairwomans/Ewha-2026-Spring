```python
import requests
import csv
from dotenv import load_dotenv
import os

load_dotenv()

TOKEN = os.getenv("GITHUB_TOKEN")
OWNER = 'ekyuho'
REPO = 'Ewha-2026-Spring'

headers = {"Authorization": f"token {TOKEN}"}
prs = []
page = 1
url = f"https://api.github.com/repos/{OWNER}/{REPO}/pulls"
print(url)

while True:
    res = requests.get(
        url,
        headers=headers,
        params={"state": "all", "per_page": 100, "page": page}
    )
    data = res.json()
    if not data:
        break
    prs.extend(data)
    page += 1

with open("pr_history.csv", "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerow(["번호", "제목", "작성자", "생성일", "병합일", "상태"])
    for pr in prs:
        writer.writerow([
            pr["number"], pr["title"], pr["user"]["login"],
            pr["created_at"], pr.get("merged_at", ""), pr["state"]
        ])

print(f"총 {len(prs)}개 PR 추출 완료 → pr_history.csv")
'''

**.env 파일:**

GITHUB_TOKEN=ghp_your_token_here
GITHUB_OWNER=your-org-or-username
GITHUB_REPO=your-repo-name
'''
```
